# Welcome to Objective-C Jupyter!

This notebook runs Objective-C code directly in your browser using WebAssembly.
No native toolchain required — everything runs client-side.

The kernel supports: NSLog, @interface/@implementation, message sends,
variable declarations, integer arithmetic, string literals, control flow
(if/else, while, for), @property/@synthesize, Foundation class stubs
(NSString, NSNumber, NSArray), and dot syntax.

```objective-c
// Hello World
NSLog(@"Hello from Objective-C in WebAssembly!");
```

```objective-c
// Variables and arithmetic
int x = 10;
int y = 20;
int sum = x + y;
NSLog(@"%d + %d = %d", x, y, sum);
```

```objective-c
// Define a class with a method
@interface Dog
- (void)bark;
@end

@implementation Dog
- (void)bark {
    NSLog(@"Woof!");
}
@end
```

```objective-c
// Use the class (state persists across cells)
Dog *d = [[Dog alloc] init];
[d bark];
```
