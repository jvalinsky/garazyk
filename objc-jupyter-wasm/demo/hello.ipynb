# Welcome to Objective-C Jupyter Kernel!

This notebook runs Objective-C code directly in your browser using WebAssembly.
No native toolchain required - everything runs client-side.

```objective-c
// Cell 1: Hello World
NSString *greeting = @"Hello from Objective-C in WebAssembly!";
NSLog(@"%@", greeting);
```

```objective-c
// Cell 2: Create a class
@interface Greeter : NSObject
- (NSString *)greet:(NSString *)name;
@end

@implementation Greeter
- (NSString *)greet:(NSString *)name {
    return [NSString stringWithFormat:@"Hello, %@!", name];
}
@end
```

```objective-c
// Cell 3: Use the class (state persists across cells)
Greeter *g = [[Greeter alloc] init];
NSString *result = [g greet:@"World"];
NSLog(@"Result: %@", result);
```

```objective-c
// Cell 4: Foundation classes
NSArray *languages = @[@"Objective-C", @"WebAssembly", @"Jupyter"];
NSLog(@"Languages: %@", languages);
NSLog(@"Count: %ld", (long)[languages count]);
```
