# Welcome to Objective-C Jupyter Kernel!

This notebook runs Objective-C code directly in your browser using WebAssembly.
No native toolchain required - everything runs client-side.

> The kernel currently supports: NSLog, @interface/@implementation, message sends,
> variable declarations, integer arithmetic, and string literals. Foundation classes
> (NSString, NSArray, etc.) are not yet available.

```objective-c
// Cell 1: Hello World
NSLog(@"Hello from Objective-C in WebAssembly!");
```

```objective-c
// Cell 2: Variables and arithmetic
int x = 10;
int y = 20;
int sum = x + y;
NSLog(@"%d + %d = %d", x, y, sum);
```

```objective-c
// Cell 3: Define a class
@interface Dog
- (void)bark;
@end

@implementation Dog
- (void)bark {
    NSLog(@"Woof!");
}
@end
```

```objective-c
// Cell 4: Use the class (state persists across cells)
Dog *d = [Dog alloc];
[d bark];
```
