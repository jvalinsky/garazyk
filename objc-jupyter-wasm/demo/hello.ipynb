# Welcome to Objective-C Jupyter!

This notebook runs Objective-C code directly in your browser using WebAssembly.
No native toolchain required — everything runs client-side.

The kernel supports: NSLog, @interface/@implementation, message sends,
variable declarations, integer and float arithmetic, string literals,
control flow (if/else, while, for, for-in), @property/@synthesize,
Foundation class stubs (NSString, NSNumber, NSArray, NSDictionary,
NSSet, NSData), blocks, dot syntax, and tab completion.

```objective-c
// Hello World
NSLog(@"Hello from Objective-C in WebAssembly!");
```

```objective-c
// Variables and arithmetic
int x = 10;
int y = 20;
int sum = x + y;
NSLog(@"%d + %d = %d", x, y, sum);
```

```objective-c
// Float arithmetic
float pi = 3.14159;
float r = 5.0;
float area = pi * r * r;
NSLog(@"Area of circle: %f", area);
```

```objective-c
// Ternary operator
int age = 20;
NSString *status = age >= 18 ? @"adult" : @"minor";
NSLog(@"Status: %@", status);
```

```objective-c
// for-in over a string (character iteration)
NSString *greeting = @"Hello";
for (NSString *c in greeting) {
    NSLog(@"char: %@", c);
}
```

```objective-c
// Define a class with a method
@interface Dog
- (void)bark;
@end

@implementation Dog
- (void)bark {
    NSLog(@"Woof!");
}
@end
```

```objective-c
// Use the class (state persists across cells)
Dog *d = [[Dog alloc] init];
[d bark];
```
