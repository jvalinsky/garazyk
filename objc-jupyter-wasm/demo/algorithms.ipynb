# Algorithms in Objective-C

Classic algorithms implemented in the ObjC interpreter.
Demonstrates loops, collections, blocks, and string manipulation.

```objective-c
// FizzBuzz: 1 to 20
for (int i = 1; i <= 20; i++) {
    if (i % 15 == 0) {
        NSLog(@"FizzBuzz");
    } else if (i % 3 == 0) {
        NSLog(@"Fizz");
    } else if (i % 5 == 0) {
        NSLog(@"Buzz");
    } else {
        NSLog(@"%d", i);
    }
}
```

```objective-c
// Fibonacci sequence (first 15 numbers)
int a = 0;
int b = 1;
for (int i = 0; i < 15; i++) {
    NSLog(@"fib(%d) = %d", i, a);
    int temp = a + b;
    a = b;
    b = temp;
}
```

```objective-c
// Sieve of Eratosthenes (primes up to 30)
// Using NSMutableArray as a boolean array (0 = composite, 1 = prime)
int limit = 30;
NSMutableArray *sieve = [NSMutableArray array];
for (int i = 0; i <= limit; i++) {
    [sieve addObject:[NSNumber numberWithInt:1]];
}
// 0 and 1 are not prime
[sieve replaceObjectAtIndex:0 withObject:[NSNumber numberWithInt:0]];
[sieve replaceObjectAtIndex:1 withObject:[NSNumber numberWithInt:0]];

for (int p = 2; p * p <= limit; p++) {
    if ([[sieve objectAtIndex:p] intValue] == 1) {
        for (int m = p * p; m <= limit; m += p) {
            [sieve replaceObjectAtIndex:m withObject:[NSNumber numberWithInt:0]];
        }
    }
}

NSLog(@"Primes up to %d:", limit);
for (int i = 2; i <= limit; i++) {
    if ([[sieve objectAtIndex:i] intValue] == 1) {
        NSLog(@"  %d", i);
    }
}
```

```objective-c
// Word frequency counter using blocks
NSString *text = @"the cat sat on the mat the cat";
NSArray *words = [text componentsSeparatedByString:@" "];

__block NSMutableDictionary *freq = [NSMutableDictionary dictionary];
[words enumerateObjectsUsingBlock:^(id word, int idx, int *stop) {
    NSNumber *count = [freq objectForKey:word];
    if (count == 0) {
        [freq setObject:[NSNumber numberWithInt:1] forKey:word];
    } else {
        [freq setObject:[NSNumber numberWithInt:[count intValue] + 1] forKey:word];
    }
}];

for (NSString *w in freq) {
    NSLog(@"%@: %d", w, [[freq objectForKey:w] intValue]);
}
```

```objective-c
// Factorial with a class method
@interface Math
+ (int)factorial:(int)n;
@end

@implementation Math
+ (int)factorial:(int)n {
    int result = 1;
    for (int i = 2; i <= n; i++) {
        result = result * i;
    }
    return result;
}
@end

NSLog(@"10! = %d", [Math factorial:10]);
NSLog(@"12! = %d", [Math factorial:12]);
```

```objective-c
// String processing: palindrome check
NSString *word = @"racecar";
NSString *lower = [word lowercaseString];
int len = [lower length];
int isPalindrome = 1;
for (int i = 0; i < len / 2; i++) {
    NSString *left = [lower substringFromIndex:i];
    NSString *right = [lower substringFromIndex:len - 1 - i];
    if ([left characterAtIndex:0] != [right characterAtIndex:0]) {
        isPalindrome = 0;
        break;
    }
}
NSLog(@"'%@' is %@", word, isPalindrome ? @"a palindrome" : @"not a palindrome");
```
