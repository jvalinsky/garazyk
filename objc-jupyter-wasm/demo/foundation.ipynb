# Foundation Classes & Properties

The kernel provides stub implementations of common Foundation classes
(NSString, NSNumber, NSArray) and supports @property/@synthesize
for custom class properties with per-instance storage.

```objective-c
// NSString operations
NSString *greeting = @"Hello";
NSLog(@"Length: %d", [greeting length]);
NSLog(@"Int value: %d", [@"42" intValue]);

NSString *full = [greeting stringByAppendingString:@" World"];
NSLog(@"Concatenated: %@", full);

BOOL same = [@"abc" isEqualToString:@"abc"];
NSLog(@"Equal: %d", same);
```

```objective-c
// NSNumber
NSNumber *num = [NSNumber numberWithInt:42];
NSLog(@"Int value: %d", [num intValue]);
NSLog(@"Bool value: %d", [num boolValue]);
```

```objective-c
// NSArray
NSArray *arr = [NSArray array];
NSLog(@"Array count: %d", [arr count]);
```

```objective-c
// @property with @synthesize
@interface Counter : NSObject
@property (nonatomic, assign) int count;
- (void)increment;
@end

@implementation Counter
@synthesize count = _count;
- (void)increment {
    _count = _count + 1;
}
@end
```

```objective-c
// Use the Counter class with dot syntax
Counter *c = [[Counter alloc] init];
[c setCount:5];
NSLog(@"Count: %d", [c count]);

[c increment];
NSLog(@"After increment: %d", [c count]);

// Dot syntax also works
c.count = 100;
NSLog(@"Dot syntax: %d", c.count);
```
