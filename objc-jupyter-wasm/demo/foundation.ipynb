# Foundation Classes, Collections & Properties

The kernel provides stub implementations of common Foundation classes:
- `NSString`, `NSNumber`, `NSArray`, `NSMutableArray`
- `NSDictionary`, `NSMutableDictionary`, `NSSet`, `NSData`
- Supports `@property`/`@synthesize` for custom class properties
- Modern literals: `@[]`, `@{}`, `@(expr)`, `@YES`/`@NO`
- Cross-cell state persistence for `@interface` declarations


## Method Dispatching

Foundation methods are dispatched via message sends:

$$
[\text{object} \text{method}:\text{arg}]
$$

The kernel intercepts known selectors and returns stub values.

```objective-c
// NSString operations
NSString *greeting = @"Hello";
NSLog(@"Length: %d", [greeting length]);
NSLog(@"Int value: %d", [@"42" intValue]);

NSString *full = [greeting stringByAppendingString:@" World"];
NSLog(@"Concatenated: %@", full);
```

```objective-c
// String comparison
BOOL same = [@"abc" isEqualToString:@"abc"];
NSLog(@"Equal: %d", same);

// New: stringByAppendingPathComponent:
NSString *path = @"/usr/local";
NSString *fullPath = [path stringByAppendingPathComponent:@"bin"];
NSLog(@"Path: %@", fullPath);

// New: capitalizedString
NSString *name = @"hello world";
NSLog(@"Capitalized: %@", [name capitalizedString]);
```

```objective-c
// NSString manipulation
NSString *sentence = @"Hello World Hello";
NSString *replaced = [sentence stringByReplacingOccurrencesOfString:@"Hello" withString:@"Hi"];
NSLog(@"Replaced: %@", replaced);

NSString *csv = @"red,green,blue";
NSArray *parts = [csv componentsSeparatedByString:@","];
NSLog(@"Split count: %d", [parts count]);

NSString *padded = @"  hello  ";
NSString *trimmed = [padded stringByTrimmingCharactersInSet:[NSCharacterSet whitespaceCharacterSet]];
NSLog(@"Trimmed: '%@'", trimmed);
```

## NSNumber Factory Methods

| Method | Description | Example |
|--------|-------------|----------|
| `numberWithInt:` | Wrap int | `@(42)` or `[NSNumber numberWithInt:42]` |
| `numberWithDouble:` | Wrap double | `@(3.14)` or `[NSNumber numberWithDouble:3.14]` |
| `numberWithBool:` | Wrap BOOL | `@(YES)` or `[NSNumber numberWithBool:YES]` |
| `numberWithUnsignedInt:` | Wrap unsigned int | `[NSNumber numberWithUnsignedInt:42]` |
| `numberWithLong:` | Wrap long | `[NSNumber numberWithLong:100]` |
| `numberWithInteger:` | Wrap NSInteger | `[NSNumber numberWithInteger:50]` |


```objective-c
// NSNumber
NSNumber *num = [NSNumber numberWithInt:42];
NSLog(@"Int value: %d", [num intValue]);
NSLog(@"Bool value: %d", [num boolValue]);

NSNumber *flag = [NSNumber numberWithBool:YES];
NSLog(@"Bool as string: %@", [flag stringValue]);

// New: numberWithUnsignedInt:, numberWithLong:, numberWithInteger:
NSNumber *u = [NSNumber numberWithUnsignedInt:42];
NSNumber *l = [NSNumber numberWithLong:100];
NSNumber *i = [NSNumber numberWithInteger:50];
NSLog(@"Unsigned: %@", u);
NSLog(@"Long: %@", l);
NSLog(@"Integer: %@", i);

// New: longLongValue, unsignedIntValue
NSNumber *n = @(42);
NSLog(@"longLongValue: %d", [n longLongValue]);
NSLog(@"unsignedIntValue: %d", [n unsignedIntValue]);
```

```objective-c
// NSData — binary data with hex encoding
NSData *empty = [NSData data];
NSLog(@"Empty data length: %d", [empty length]);

NSData *data = [NSData dataWithBytes:@"AB" length:2];
NSLog(@"Data length: %d", [data length]);
NSLog(@"Data description: %@", [data description]);
```

## Collections: NSArray / NSMutableArray

$$
\text{NSArray} \xrightarrow{\text{mutable}} \text{NSMutableArray}
$$


```objective-c
// NSMutableArray — add, access, remove
NSMutableArray *arr = [NSMutableArray array];
[arr addObject:@"hello"];
[arr addObject:@"world"];
NSLog(@"Count: %d", [arr count]);
NSLog(@"First: %@", [arr objectAtIndex:0]);
NSLog(@"Last: %@", [arr lastObject]);

// New: firstObject
NSLog(@"First (again): %@", [arr firstObject]);

// New: arrayByAddingObject:
NSArray *extended = [arr arrayByAddingObject:@"extra"];
NSLog(@"Extended: %@", extended);

[arr removeLastObject];
NSLog(@"After remove: %d items, last=%@", [arr count], [arr lastObject]);
```

```objective-c
// NSArray iteration
NSArray *colors = @[@"red", @"green", @"blue"];
for (NSString *c in colors) {
    NSLog(@"color: %@", c);
}

// Check containment
BOOL hasRed = [colors containsObject:@"red"];
NSLog(@"Has red: %d", hasRed);
```

## Collections: NSDictionary / NSMutableDictionary

| Method | Description |
|--------|-------------|
| `setObject:forKey:` | Add/udpate entry |
| `objectForKey:` | Get value by key |
| `removeObjectForKey:` | Remove entry |
| `allKeys` / `allValues` | Get keys/values as array |


```objective-c
// NSMutableDictionary — set, get, remove
NSMutableDictionary *dict = [NSMutableDictionary dictionary];
[dict setObject:@"apple" forKey:@"fruit"];
[dict setObject:@"carrot" forKey:@"veg"];
NSLog(@"Count: %d", [dict count]);
NSLog(@"Fruit: %@", [dict objectForKey:@"fruit"]);

// setValue:forKey: also works
[dict setValue:@"banana" forKey:@"fruit"];
NSLog(@"Updated: %@", [dict valueForKey:@"fruit"]);

[dict removeObjectForKey:@"veg"];
NSLog(@"After remove: %d", [dict count]);
```

```objective-c
// NSDictionary creation and equality
NSDictionary *d1 = [NSDictionary dictionaryWithObject:@"value" forKey:@"key"];
NSDictionary *d2 = [NSDictionary dictionaryWithObject:@"value" forKey:@"key"];
NSLog(@"Equal: %d", [d1 isEqualToDictionary:d2]);

// New: stringWithString:
NSString *original = @"test";
NSString *copy = [NSString stringWithString:original];
NSLog(@"Copy: %@", copy);
```

```objective-c
// for-in over collections
NSMutableArray *colors2 = [NSMutableArray array];
[colors2 addObject:@"red"];
[colors2 addObject:@"green"];
[colors2 addObject:@"blue"];
for (NSString *c in colors2) {
    NSLog(@"color: %@", c);
}

// for-in over dict keys
NSMutableDictionary *d = [NSMutableDictionary dictionary];
[d setObject:@"1" forKey:@"a"];
[d setObject:@"2" forKey:@"b"];
for (NSString *k in d) {
    NSLog(@"%@ = %@", k, [d objectForKey:k]);
}
```

## NSSet — Unique Elements

$$
\text{NSArray} \xrightarrow{\text{unique}} \text{NSSet}
$$


```objective-c
// NSSet — unique elements
NSMutableArray *src = [NSMutableArray array];
[src addObject:@"a"];
[src addObject:@"b"];
[src addObject:@"a"];
NSSet *set = [NSSet setWithArray:src];
NSLog(@"Set count: %d (deduped from 3)", [set count]);
NSLog(@"Contains a: %d", [set containsObject:@"a"]);
NSLog(@"Contains c: %d", [set containsObject:@"c"]);
```

## @property and @synthesize

$$
\begin{aligned}
&@\text{interface} \text{ClassName} : \text{SuperClass} \\\\n&\quad @\text{property} (\text{attributes}) \text{Type} *\text{name}; \\\\n&@\text{end}
\\
&@\text{implementation} \text{ClassName} \\\\n&\quad @\text{synthesize} \text{name} = \_\text{name}; \\\\n&@\text{end}
\end{aligned}
$$


```objective-c
// @property with @synthesize
@interface Counter : NSObject
@property (nonatomic, assign) int count;
- (void)increment;
@end

@implementation Counter
@synthesize count = _count;
- (void)increment {
    _count = _count + 1;
}
@end

// Use dot syntax
Counter *c = [[Counter alloc] init];
[c setCount:5];
NSLog(@"Count: %d", [c count]);

c.count = 100;
NSLog(@"Dot syntax: %d", c.count);
```
