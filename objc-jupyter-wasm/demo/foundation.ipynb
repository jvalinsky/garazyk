# Objective-C Runtime in WebAssembly

This notebook demonstrates the Objective-C runtime features available
in the browser-based kernel. The kernel uses the real GNUstep libobjc2
runtime compiled to WebAssembly.

> Foundation classes (NSString, NSArray, NSDictionary) are not yet available.
> The kernel currently supports: NSLog, @interface/@implementation, message sends,
> variable declarations, integer arithmetic, and string literals.

```objective-c
// NSLog with format strings
NSLog(@"Hello from the ObjC runtime!");
NSLog(@"Integer: %d", 42);
NSLog(@"Percent: %%");
```

```objective-c
// Define a custom class
@interface Calculator
- (int)add:(int)a to:(int)b;
@end

@implementation Calculator
- (int)add:(int)a to:(int)b {
    return a + b;
}
@end
```

```objective-c
// Use the Calculator class
Calculator *calc = [Calculator alloc];
int result = [calc add:3 to:4];
NSLog(@"3 + 4 = %d", result);
```
