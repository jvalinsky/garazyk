# Objective-C Foundation in WebAssembly

> **Note:** This is a target-state demo. The current smoke kernel echoes
> submitted code without executing it. Foundation class support is a
> planned follow-up layer.

This notebook will demonstrate using Foundation classes in the browser-based kernel.

```objective-c
// NSString examples
NSString *str1 = @"Hello";
NSString *str2 = @"World";
NSString *combined = [str1 stringByAppendingString:str2];
NSLog(@"%@", combined);
```

```objective-c
// NSArray examples
NSArray *array = @[@"one", @"two", @"three"];
for (NSString *item in array) {
    NSLog(@"Item: %@", item);
}
```

```objective-c
// NSDictionary examples
NSDictionary *dict = @{
    @"name": @"Objective-C",
    @"runtime": @"WebAssembly",
    @"kernel": @"Jupyter"
};

NSLog(@"Dict: %@", dict);
NSLog(@"Name: %@", dict[@"name"]);
```
