# Foundation Classes, Collections & Properties

The kernel provides stub implementations of common Foundation classes
(NSString, NSNumber, NSArray, NSMutableArray, NSDictionary, NSMutableDictionary,
NSSet, NSData) and supports @property/@synthesize for custom class properties
with per-instance storage.

```objective-c
// NSString operations
NSString *greeting = @"Hello";
NSLog(@"Length: %d", [greeting length]);
NSLog(@"Int value: %d", [@"42" intValue]);

NSString *full = [greeting stringByAppendingString:@" World"];
NSLog(@"Concatenated: %@", full);

BOOL same = [@"abc" isEqualToString:@"abc"];
NSLog(@"Equal: %d", same);
```

```objective-c
// NSString manipulation
NSString *sentence = @"Hello World Hello";
NSString *replaced = [sentence stringByReplacingOccurrencesOfString:@"Hello" withString:@"Hi"];
NSLog(@"Replaced: %@", replaced);

NSString *csv = @"red,green,blue";
NSArray *parts = [csv componentsSeparatedByString:@","];
NSLog(@"Split count: %d", [parts count]);

NSString *padded = @"  hello  ";
NSString *trimmed = [padded stringByTrimmingWhitespace];
NSLog(@"Trimmed: '%@'", trimmed);
```

```objective-c
// NSNumber
NSNumber *num = [NSNumber numberWithInt:42];
NSLog(@"Int value: %d", [num intValue]);
NSLog(@"Bool value: %d", [num boolValue]);

NSNumber *flag = [NSNumber numberWithBool:YES];
NSLog(@"Bool as string: %@", [flag stringValue]);
```

```objective-c
// NSData — binary data with hex encoding
NSData *empty = [NSData data];
NSLog(@"Empty data length: %d", [empty length]);

NSData *data = [NSData dataWithBytes:@"AB" length:2];
NSLog(@"Data length: %d", [data length]);
NSLog(@"Data description: %@", [data description]);
```

```objective-c
// NSMutableArray — add, access, remove
NSMutableArray *arr = [NSMutableArray array];
[arr addObject:@"hello"];
[arr addObject:@"world"];
NSLog(@"Count: %d", [arr count]);
NSLog(@"First: %@", [arr objectAtIndex:0]);
NSLog(@"Last: %@", [arr lastObject]);

[arr removeLastObject];
NSLog(@"After remove: %d items, last=%@", [arr count], [arr lastObject]);
```

```objective-c
// NSMutableDictionary — set, get, remove
NSMutableDictionary *dict = [NSMutableDictionary dictionary];
[dict setObject:@"apple" forKey:@"fruit"];
[dict setObject:@"carrot" forKey:@"veg"];
NSLog(@"Count: %d", [dict count]);
NSLog(@"Fruit: %@", [dict objectForKey:@"fruit"]);

// setValue:forKey: also works
[dict setValue:@"banana" forKey:@"fruit"];
NSLog(@"Updated: %@", [dict valueForKey:@"fruit"]);

[dict removeObjectForKey:@"veg"];
NSLog(@"After remove: %d", [dict count]);
```

```objective-c
// NSDictionary creation and equality
NSDictionary *d1 = [NSDictionary dictionaryWithObject:@"value" forKey:@"key"];
NSDictionary *d2 = [NSDictionary dictionaryWithObject:@"value" forKey:@"key"];
NSLog(@"Equal: %d", [d1 isEqualToDictionary:d2]);
```

```objective-c
// for-in over collections
NSMutableArray *colors = [NSMutableArray array];
[colors addObject:@"red"];
[colors addObject:@"green"];
[colors addObject:@"blue"];
for (NSString *c in colors) {
    NSLog(@"color: %@", c);
}

// for-in over dict keys
NSMutableDictionary *d = [NSMutableDictionary dictionary];
[d setObject:@"1" forKey:@"a"];
[d setObject:@"2" forKey:@"b"];
for (NSString *k in d) {
    NSLog(@"%@ = %@", k, [d objectForKey:k]);
}
```

```objective-c
// NSSet — unique elements
NSMutableArray *src = [NSMutableArray array];
[src addObject:@"a"];
[src addObject:@"b"];
[src addObject:@"a"];
NSSet *set = [NSSet setWithArray:src];
NSLog(@"Set count: %d (deduped from 3)", [set count]);
NSLog(@"Contains a: %d", [set containsObject:@"a"]);
NSLog(@"Contains c: %d", [set containsObject:@"c"]);
```

```objective-c
// @property with @synthesize
@interface Counter : NSObject
@property (nonatomic, assign) int count;
- (void)increment;
@end

@implementation Counter
@synthesize count = _count;
- (void)increment {
    _count = _count + 1;
}
@end
```

```objective-c
// Use the Counter class with dot syntax
Counter *c = [[Counter alloc] init];
[c setCount:5];
NSLog(@"Count: %d", [c count]);

[c increment];
NSLog(@"After increment: %d", [c count]);

// Dot syntax also works
c.count = 100;
NSLog(@"Dot syntax: %d", c.count);
```
