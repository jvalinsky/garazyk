## How This Works

Every notebook runs Objective-C code in a WebAssembly interpreter. The kernel
parses ObjC source, executes it, and returns results — all client-side.

No native toolchain, no Xcode, no compiler. Just a browser tab.

The kernel supports:
- `@interface` / `@implementation`, `@property`, `@synthesize`
- Message sends, multi-keyword selectors, `nil` messaging
- Blocks with `__block` capture, `enumerateObjectsUsingBlock:`
- `@try` / `@catch` / `@finally` / `@throw`
- `@protocol`, `@autoreleasepool`, categories
- Foundation: NSString, NSNumber, NSArray, NSDictionary, NSSet, NSData
- Literals: `@[]`, `@{}`, `@(expr)`, `@YES`/`@NO`

```objective-c
// Your first cell — try running this
NSLog(@"Hello from the ObjC WASM kernel!");

NSArray *parts = @[@"at://", @"did:plc:", @"abc"];
NSLog(@"DID: %@", [parts componentsJoinedByString:@""]);
```

## Support Classes

Every code cell in this course falls into one of four categories:

| Class | Description | Example |
|-------|-------------|----------|
| **direct** | Runs in the interpreter without host services | Classes, properties, blocks, Foundation collections |
| **host-bridge** | Uses deterministic browser or JS host services | JSON parsing, SHA-256, Base32 encoding |
| **adapted-shim** | Mirrors a production boundary with an in-memory model | Repository storage, blob metadata, XRPC dispatch |
| **unsupported-production** | Production APIs that cannot run in the browser | SQLite, GCD, filesystem, Security framework |

When a notebook uses an **adapted-shim**, the code structure mirrors production
but replaces real I/O with in-memory dictionaries. When it uses a **host-bridge**,
the kernel calls out to JavaScript for crypto or encoding operations.

**unsupported-production** cells are never sent to the interpreter. They appear
as diagnostics in the test runner, not as runtime errors.

## How to Use These Notebooks

1. **Run cells in order.** State persists across cells within a notebook.
2. **Read the markdown.** Each step explains the concept before the code.
3. **Try the exercises.** Placeholder cells (`// Your code here...`) are where you practice.
4. **Check the solutions.** Solution cells follow each exercise.
5. **Follow the chain.** Each notebook ends with "What to Read Next" links.

The kernel resets between notebooks — classes defined in one notebook are not
available in the next. Each notebook is self-contained.
