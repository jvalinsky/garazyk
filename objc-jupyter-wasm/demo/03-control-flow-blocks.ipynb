## Classes and Methods

```objective-c
// Control flow: while loop
int i = 1;
int total = 0;
while (i <= 10) {
    total += i;
    i++;
}
NSLog(@"Sum 1..10 = %d", total);
```

```objective-c
// Control flow: for loop
int factorial = 1;
for (int n = 1; n <= 7; n++) {
    factorial = factorial * n;
}
NSLog(@"7! = %d", factorial);
```

```objective-c
// Define a class with a method that returns a value
@interface Calculator : NSObject
- (int)add:(int)a to:(int)b;
+ (Calculator *)sharedInstance;
@end

static Calculator *sharedCalc = 0;

@implementation Calculator
- (int)add:(int)a to:(int)b {
    return a + b;
}
+ (Calculator *)sharedInstance {
    if (!sharedCalc) {
        sharedCalc = [[Calculator alloc] init];
    }
    return sharedCalc;
}
@end
```

```objective-c
// Use the Calculator: nested messages and class methods
Calculator *calc = [[Calculator alloc] init];
int result = [calc add:3 to:4];
NSLog(@"3 + 4 = %d", result);

// Class method
Calculator *shared = [Calculator sharedInstance];
NSLog(@"Shared: %d", [shared add:10 to:20]);
```

## Blocks: Anonymous Functions

Blocks are closures—functions you can define inline and pass to other methods.

```objective-c
// Blocks — anonymous functions
int multiplier = 3;
void (^printDoubled)(int) = ^(int n) {
    NSLog(@"%d * %d = %d", n, multiplier, n * multiplier);
};
printDoubled(5);
printDoubled(10);
```

```objective-c
// enumerateObjectsUsingBlock: with stop flag
NSMutableArray *nums = [NSMutableArray array];
[nums addObject:@"one"];
[nums addObject:@"two"];
[nums addObject:@"three"];
[nums addObject:@"four"];

[nums enumerateObjectsUsingBlock:^(id obj, int idx, int *stop) {
    NSLog(@"%d: %@", idx, obj);
    if (idx == 1) {
        *stop = 1;
    }
}];
```

```objective-c
// __block capture — mutations visible outside
__block int sum = 0;
NSMutableArray *items = [NSMutableArray array];
[items addObject:@"a"];
[items addObject:@"b"];
[items addObject:@"c"];

[items enumerateObjectsUsingBlock:^(id obj, int idx, int *stop) {
    sum = sum + idx;
}];
NSLog(@"Sum of indices: %d", sum);
```

## Collections: Building a Word Frequency Counter

Combine dictionaries and enumeration to track data across a dataset.

```objective-c
// Collections: building a word frequency counter
NSMutableArray *words = [NSMutableArray array];
[words addObject:@"hello"];
[words addObject:@"world"];
[words addObject:@"hello"];
[words addObject:@"objc"];

NSMutableDictionary *freq = [NSMutableDictionary dictionary];
for (NSString *w in words) {
    NSNumber *n = [freq objectForKey:w];
    if (n == 0 || [n intValue] == 0) {
        [freq setObject:[NSNumber numberWithInt:1] forKey:w];
    } else {
        [freq setObject:[NSNumber numberWithInt:[n intValue] + 1] forKey:w];
    }
}

for (NSString *k in freq) {
    NSLog(@"%@: %d", k, [[freq objectForKey:k] intValue]);
}
```

## Algorithm Practice

These classic algorithms reinforce the control flow and collection patterns above.

```objective-c
// FizzBuzz: 1 to 20
for (int i = 1; i <= 20; i++) {
    if (i % 15 == 0) {
        NSLog(@"FizzBuzz");
    } else if (i % 3 == 0) {
        NSLog(@"Fizz");
    } else if (i % 5 == 0) {
        NSLog(@"Buzz");
    } else {
        NSLog(@"%d", i);
    }
}
```

```objective-c
// Fibonacci sequence (first 15 numbers)
int a = 0;
int b = 1;
for (int i = 0; i < 15; i++) {
    NSLog(@"fib(%d) = %d", i, a);
    int temp = a + b;
    a = b;
    b = temp;
}
```

```objective-c
// Sieve of Eratosthenes (primes up to 30)
// Using NSMutableArray as a boolean array (0 = composite, 1 = prime)
int limit = 30;
NSMutableArray *sieve = [NSMutableArray array];
for (int i = 0; i <= limit; i++) {
    [sieve addObject:[NSNumber numberWithInt:1]];
}
// 0 and 1 are not prime
[sieve replaceObjectAtIndex:0 withObject:[NSNumber numberWithInt:0]];
[sieve replaceObjectAtIndex:1 withObject:[NSNumber numberWithInt:0]];

for (int p = 2; p * p <= limit; p++) {
    if ([[sieve objectAtIndex:p] intValue] == 1) {
        for (int m = p * p; m <= limit; m += p) {
            [sieve replaceObjectAtIndex:m withObject:[NSNumber numberWithInt:0]];
        }
    }
}

NSLog(@"Primes up to %d:", limit);
for (int i = 2; i <= limit; i++) {
    if ([[sieve objectAtIndex:i] intValue] == 1) {
        NSLog(@"  %d", i);
    }
}
```

```objective-c
// Word frequency counter using blocks
NSString *text = @"the cat sat on the mat the cat";
NSArray *words = [text componentsSeparatedByString:@" "];

__block NSMutableDictionary *freq = [NSMutableDictionary dictionary];
[words enumerateObjectsUsingBlock:^(id word, int idx, int *stop) {
    NSNumber *count = [freq objectForKey:word];
    if (count == 0) {
        [freq setObject:[NSNumber numberWithInt:1] forKey:word];
    } else {
        [freq setObject:[NSNumber numberWithInt:[count intValue] + 1] forKey:word];
    }
}];

for (NSString *w in freq) {
    NSLog(@"%@: %d", w, [[freq objectForKey:w] intValue]);
}
```

```objective-c
// Factorial with a class method
@interface Math
+ (int)factorial:(int)n;
@end

@implementation Math
+ (int)factorial:(int)n {
    int result = 1;
    for (int i = 2; i <= n; i++) {
        result = result * i;
    }
    return result;
}
@end

NSLog(@"10! = %d", [Math factorial:10]);
NSLog(@"12! = %d", [Math factorial:12]);
```

```objective-c
// String processing: palindrome check
NSString *word = @"racecar";
NSString *lower = [word lowercaseString];
int len = [lower length];
int isPalindrome = 1;
for (int i = 0; i < len / 2; i++) {
    NSString *left = [lower substringFromIndex:i];
    NSString *right = [lower substringFromIndex:len - 1 - i];
    if ([left characterAtIndex:0] != [right characterAtIndex:0]) {
        isPalindrome = 0;
        break;
    }
}
NSLog(@"'%@' is %@", word, isPalindrome ? @"a palindrome" : @"not a palindrome");
```
