## Basic Syntax

Objective-C uses message sends instead of function calls:

$$
\text{[receiver method:arg]}
$$


```objective-c
// Hello World
NSLog(@"Hello from Objective-C in WebAssembly!");
```

```objective-c
// Variables and arithmetic
int x = 10;
int y = 20;
int sum = x + y;
NSLog(@"%d + %d = %d", x, y, sum);
```

```objective-c
// Float arithmetic
float pi = 3.14159;
float r = 5.0;
float area = pi * r * r;
NSLog(@"Area of circle: %f", area);
```

## Ternary Operator

$$
\text{condition} ? \text{true_value} : \text{false_value}
$$


```objective-c
// Ternary operator
int age = 20;
NSString *status = age >= 18 ? @"adult" : @"minor";
NSLog(@"Status: %@", status);
```

```objective-c
// for-in over a string (character iteration)
NSString *greeting = @"Hello";
for (NSString *c in greeting) {
    NSLog(@"char: %@", c);
}
```

## Classes and Methods

Define classes with `@interface` and `@implementation`:

$$
\begin{aligned}
&@\text{interface ClassName : SuperClass} \\\\n&\quad \text{// ivars and properties} \\\\n&\quad - (\text{return_type})\text{method}; \\\\n&@\text{end}
\end{aligned}
$$


```objective-c
// Define a class
@interface Dog : NSObject
- (void)bark;
@end

@implementation Dog
- (void)bark {
    NSLog(@"Woof!");
}
@end

// Use the class
Dog *d = [[Dog alloc] init];
[d bark];
```

## nil Checking

`nil` is a first-class value. All these patterns work:

| Pattern      | Equivalent To     |
| ------------ | ----------------- |
| `if (obj)`   | `if (obj != nil)` |
| `if (!obj)`  | `if (obj == nil)` |
| `obj == nil` | `obj == (id)0`    |
| `obj != nil` | `obj != (id)0`    |

```objective-c
// nil checking patterns
NSString *maybeNil = nil;
if (!maybeNil) {
    NSLog(@"maybeNil is nil");
}

if (maybeNil == nil) {
    NSLog(@"Confirmed: it's nil");
}

maybeNil = @"Hello";
if (maybeNil) {
    NSLog(@"Now it's not nil: %@", maybeNil);
}
```

## Modern Literals

| Literal      | Description        | Example           |
| ------------ | ------------------ | ----------------- |
| `@[]`        | Array literal      | `@[@1, @2, @3]`   |
| `@{}`        | Dictionary literal | `@{@"a": @1}`     |
| `@(expr)`    | Boxed expression   | `@(x)`, `@(3.14)` |
| `@YES`/`@NO` | Boolean literals   | `BOOL b = @YES;`  |

```objective-c
// Modern array and dictionary literals
NSArray *arr = @[@1, @2, @3];
NSLog(@"Array: %@", arr);
NSLog(@"Count: %d", (int)[arr count]);

NSDictionary *dict = @{@"key1": @"value1", @"key2": @"value2"};
NSLog(@"Dictionary: %@", dict);
```

```objective-c
// Boxed expressions
int x = 42;
float pi = 3.14159;
NSNumber *boxedInt = @(x);
NSNumber *boxedFloat = @(pi);
NSNumber *boxedYes = @(YES);
NSLog(@"Boxed int: %@", boxedInt);
NSLog(@"Boxed float: %@", boxedFloat);
NSLog(@"Boxed YES: %@", boxedYes);
```
