## Architecture

$$
\text{Client} \xrightarrow{\text{XRPC}} \text{PDSRecordService} \xrightarrow{\text{Store}} \text{Actor DB} \xrightarrow{\text{Commit}} \text{Repository}
$$


```objective-c
// XRPC method entry simulation
NSString *collection = @"app.bsky.feed.post";
NSLog(@"Collection: %@", collection);

// Record service behavior
NSDictionary *record = @{@"text": @"Hello World", @"createdAt": @"2024-01-01"};
NSLog(@"Record: %@", record);
```

```objective-c
// Actor store and repository state
NSMutableArray *commits = [NSMutableArray array];
[commits addObject:@"commit1"];
[commits addObject:@"commit2"];
NSLog(@"Commits: %@", commits);

// Commit generation and sync
NSLog(@"Repository updated");
```
