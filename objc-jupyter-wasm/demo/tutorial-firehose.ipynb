## Architecture

$$
\text{Client} \xrightarrow{\text{Upgrade}} \text{WebSocket} \xrightarrow{\text{Handler}} \text{SubscribeRepos} \xrightarrow{\text{Seq}} \text{Events}
$$


```objective-c
// WebSocket upgrade path concepts
NSLog(@"WebSocket upgrade requested");
BOOL upgraded = YES;
NSLog(@"Upgraded: %d", upgraded);
```

```objective-c
// subscribeRepos semantics
NSArray *events = @[@"commit", @"identity", @"account"];
for (NSString *event in events) {
    NSLog(@"Event: %@", event);
}

// Event ordering and replay
NSLog(@"Event ordering: FIFO");
```
