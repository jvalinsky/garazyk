# Interactive Objective-C Session

This notebook demonstrates multi-cell stateful Objective-C programming.
Variables and class definitions persist across cells, just like a real REPL.

> The kernel currently supports: NSLog, @interface/@implementation, message sends,
> variable declarations, integer arithmetic, and string literals.

```objective-c
// Cell 1: Set up a counter
int counter = 0;
counter = counter + 1;
NSLog(@"Counter: %d", counter);
```

```objective-c
// Cell 2: Counter persists from previous cell!
counter = counter + 5;
NSLog(@"Counter after increment: %d", counter);
```

```objective-c
// Cell 3: Define a class
@interface Counter
- (void)increment;
- (int)value;
@end

@implementation Counter
- (void)increment {
    NSLog(@"incremented!");
}
- (int)value {
    return 0;
}
@end
```

```objective-c
// Cell 4: Use the class from previous cell
Counter *c = [Counter alloc];
[c increment];
NSLog(@"Counter value: %d", [c value]);
```
