# Interactive Objective-C Session

> **Note:** This is a target-state demo. The current smoke kernel echoes
> submitted code without executing it. Cross-cell state persistence
> and object introspection are planned follow-up layers.

This notebook will demonstrate multi-cell stateful Objective-C programming.
Variables and objects will persist across cells, just like a real REPL.

```objective-c
// Cell 1: Set up a counter
static NSInteger counter = 0;
counter++;
NSLog(@"Counter: %ld", (long)counter);
```

```objective-c
// Cell 2: Counter persists from previous cell!
counter += 5;
NSLog(@"Counter after increment: %ld", (long)counter);
```

```objective-c
// Cell 3: Create an object (persists across cells)
@interface Person : NSObject
@property (nonatomic, strong) NSString *name;
@property (nonatomic, assign) NSInteger age;
- (NSString *)description;
@end

@implementation Person
- (NSString *)description {
    return [NSString stringWithFormat:@"Person: %@, age %ld", _name, (long)_age];
}
@end

Person *john = [[Person alloc] init];
john.name = @"John";
john.age = 30;
NSLog(@"Created: %@", john);
```

```objective-c
// Cell 4: Use the object from previous cell
john.age += 1;
NSLog(@"Updated: %@", john);
NSLog(@"Name: %@", john.name);
```
