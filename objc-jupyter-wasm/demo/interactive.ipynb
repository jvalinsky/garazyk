# Interactive Objective-C Session

This notebook demonstrates multi-cell stateful Objective-C programming.
Variables and class definitions persist across cells, just like a real REPL.

Features shown: control flow, cross-cell state, nested messages,
class methods, blocks, __block capture, and error handling.

```objective-c
// Control flow: if/else
int score = 85;
if (score >= 90) {
    NSLog(@"Grade: A");
} else if (score >= 80) {
    NSLog(@"Grade: B");
} else {
    NSLog(@"Grade: C");
}
```

```objective-c
// Control flow: while loop
int i = 1;
int total = 0;
while (i <= 10) {
    total += i;
    i++;
}
NSLog(@"Sum 1..10 = %d", total);
```

```objective-c
// Control flow: for loop
int factorial = 1;
for (int n = 1; n <= 7; n++) {
    factorial = factorial * n;
}
NSLog(@"7! = %d", factorial);
```

```objective-c
// Define a class with a method that returns a value
@interface Calculator : NSObject
- (int)add:(int)a to:(int)b;
+ (Calculator *)sharedInstance;
@end

static Calculator *sharedCalc = 0;

@implementation Calculator
- (int)add:(int)a to:(int)b {
    return a + b;
}
+ (Calculator *)sharedInstance {
    if (!sharedCalc) {
        sharedCalc = [[Calculator alloc] init];
    }
    return sharedCalc;
}
@end
```

```objective-c
// Use the Calculator: nested messages and class methods
Calculator *calc = [[Calculator alloc] init];
int result = [calc add:3 to:4];
NSLog(@"3 + 4 = %d", result);

// Class method
Calculator *shared = [Calculator sharedInstance];
NSLog(@"Shared: %d", [shared add:10 to:20]);
```

```objective-c
// Blocks — anonymous functions
int multiplier = 3;
void (^printDoubled)(int) = ^(int n) {
    NSLog(@"%d * %d = %d", n, multiplier, n * multiplier);
};
printDoubled(5);
printDoubled(10);
```

```objective-c
// enumerateObjectsUsingBlock: with stop flag
NSMutableArray *nums = [NSMutableArray array];
[nums addObject:@"one"];
[nums addObject:@"two"];
[nums addObject:@"three"];
[nums addObject:@"four"];

[nums enumerateObjectsUsingBlock:^(id obj, int idx, int *stop) {
    NSLog(@"%d: %@", idx, obj);
    if (idx == 1) {
        *stop = 1;
    }
}];
```

```objective-c
// __block capture — mutations visible outside
__block int sum = 0;
NSMutableArray *items = [NSMutableArray array];
[items addObject:@"a"];
[items addObject:@"b"];
[items addObject:@"c"];

[items enumerateObjectsUsingBlock:^(id obj, int idx, int *stop) {
    sum = sum + idx;
}];
NSLog(@"Sum of indices: %d", sum);
```

```objective-c
// Logical operators and boolean values
int x = 5;
if (x > 0 && x < 10) {
    NSLog(@"x is between 0 and 10");
}

BOOL isValid = YES;
if (!isValid) {
    NSLog(@"Invalid");
} else {
    NSLog(@"Valid!");
}
```

```objective-c
// Collections: building a word frequency counter
NSMutableArray *words = [NSMutableArray array];
[words addObject:@"hello"];
[words addObject:@"world"];
[words addObject:@"hello"];
[words addObject:@"objc"];

NSMutableDictionary *freq = [NSMutableDictionary dictionary];
for (NSString *w in words) {
    NSNumber *n = [freq objectForKey:w];
    if (n == 0 || [n intValue] == 0) {
        [freq setObject:[NSNumber numberWithInt:1] forKey:w];
    } else {
        [freq setObject:[NSNumber numberWithInt:[n intValue] + 1] forKey:w];
    }
}

for (NSString *k in freq) {
    NSLog(@"%@: %d", k, [[freq objectForKey:k] intValue]);
}
```
