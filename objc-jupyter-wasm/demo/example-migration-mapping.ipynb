# Migration Mapping Example

This notebook demonstrates how to generate audit trails for file moves.
Based on `scripts/docs/examples/migration-mapping-example.js`.

**Learning Objectives:**
- Understand what migration mapping does
- Generate single mapping entries
- Generate batch mappings
- Read and lookup mappings

**Estimated Time:** 20-30 minutes


## Batch Processing

$$
\text{File List} \xrightarrow{\text{generate}} \text{Mapping} \xrightarrow{\text{write}} \text{JSON}\n$$


```objective-c
// Example 1: Generate a single mapping entry
// Equivalent to: generateMappingEntry(source, destination, cwd)
NSDictionary *entry = @{@"source": @"plan/old.md", @"destination": @"docs/new.md", @"status": @"moved"};
NSLog(@"Mapping entry: %@", entry);

// Example 2: Generate mapping for multiple files
NSArray *fileList = @[
    @{@"source": @"plan/a.md", @"dest": @"docs/a.md"},
    @{@"source": @"plan/b.md", @"dest": @"docs/b.md"}
];
NSLog(@"Total files: %d", (int)[fileList count]);

// Equivalent to: generateMigrationMapping(fileList, cwd)
NSLog(@"Batch mapping generated");
```

```objective-c
// Lookup new path from old path
// Equivalent to: lookupNewPath(oldPath)
NSString *oldPath = @"plan/oauth2.md";
NSLog(@"Old: %@ -> New: docs/oauth2-implementation.md", oldPath);

// Read mapping
// Equivalent to: readMigrationMapping()
NSLog(@"Mapping file read");
```

## Output Format

| Field | Type | Description |
|-------|------|-------------|
| `source` | `NSString *` | Original file path |
| `destination` | `NSString *` | New file path |
| `status` | `NSString *` | `moved`, `copied`, `deleted` |
| `generatedAt` | `NSDate *` | ISO timestamp |
