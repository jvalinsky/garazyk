## ATProto URI Format

$$
\text{at://}\underbrace{\text{authority}}_{\text{host}}\underbrace{/\text{collection}}_{\text{nsid}}\underbrace{/\text{rkey}}_{\text{record key}}
$$

Example: `at://user.bsky.social/app.bsky.feed.post/3k...


```objective-c
// Parse at:// URIs
NSString *uri = @"at://user.bsky.social/app.bsky.feed.post/3kabc...";
NSLog(@"URI: %@", uri);

// Extract components using string methods
NSArray *parts = [uri componentsSeparatedByString:@"://"];
NSString *authority = [parts objectAtIndex:1];
NSArray *authParts = [authority componentsSeparatedByString:@"/"];
NSLog(@"Authority: %@", [authParts objectAtIndex:0]);
NSLog(@"Path: %@", [authParts objectAtIndex:1]);
```

```objective-c
// Extract mentions from post text
NSString *post = @"Hello @user.bsky.social! Check out @alice.bsky.social";
NSLog(@"Post: %@", post);

// Simplified mention extraction
if ([post hasPrefix:@"@"]) {
    NSLog(@"Contains mentions");
}

// Check for mention pattern
NSRange range = {0, [post length]};
NSLog(@"Post length: %d", (int)range.length);
```

```objective-c
// Validate CID format (simplified)
NSString *cid = @"bafyreia123...";
BOOL valid = [cid hasPrefix:@"bafyre"];
NSLog(@"CID valid: %d", valid);

// Link validation
BOOL isAtURI = [uri hasPrefix:@"at://"];
NSLog(@"Is at:// URI: %d", isAtURI);
```
