# Content Updater Example

This notebook demonstrates record content updates.
Based on `scripts/docs/examples/content-updater-example.js`.

**Learning Objectives:**
- Update record content in ATProto
- Handle record validation
- Use path resolution for record fields

**Estimated Time:** 20-30 minutes


## Record Update Flow

$$
\text{Record} \xrightarrow{\text{validate}} \text{Update} \xrightarrow{\text{commit}} \text{Repository}
$$


```objective-c
// Update record content
NSMutableDictionary *record = [NSMutableDictionary dictionary];
[record setObject:@"Hello World" forKey:@"text"];
[record setObject:@"2024-01-01T00:00:00Z" forKey:@"createdAt"];
NSLog(@"Record: %@", record);

// Validate update
BOOL valid = (record != nil);
NSLog(@"Update valid: %d", valid);
```

```objective-c
// Path resolver example
NSString *path = @"record.text";
NSArray *parts = [path componentsSeparatedByString:@"."];
NSLog(@"Path parts: %@", parts);

// Simulate path resolution
NSString *value = [record objectForKey:[parts objectAtIndex:1]];
NSLog(@"Resolved value: %@", value);
```

## Field Types

| Field | ATProto Type | Objective-C Type |
|-------|---------------|------------------|
| `text` | `string` | `NSString *` |
| `createdAt` | `datetime` | `NSString *` |
| `embed` | `record` | `NSDictionary *` |
| `facets` | `array` | `NSArray *` |
