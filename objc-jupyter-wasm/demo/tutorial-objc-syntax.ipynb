## Message Sends and Selectors

Objective-C uses message sends instead of function calls:

$$[\text{receiver}\;\text{method}:\text{arg}]$$

The trailing colon in `handleLookup:` means the method accepts one argument.

```objective-c
// Basic Objective-C syntax
@interface MyClass : NSObject
- (void)doSomething;
@end

@implementation MyClass
- (void)doSomething {
    NSLog(@"Hello from MyClass");
}
@end

// Create instance
MyClass *obj = [[MyClass alloc] init];
[obj doSomething];
```

```objective-c
// Message sends with arguments
NSString *str = @"Hello";
NSLog(@"Length: %d", [str length]);

// Selector with argument
BOOL equal = [str isEqualToString:@"Hello"];
NSLog(@"Equal: %d", equal);

// @selector()
SEL sel = @selector(length);
NSLog(@"Selector: %p", sel);
```

## Delegates and Data Sources

Garazyk uses Cocoa-style delegation patterns:

$$(\text{delegate})\;\xrightarrow{\text{callback}}(\text{dataSource})$$


```objective-c
// Delegate pattern (simplified)
@interface MyDelegate : NSObject
- (void)tableViewDidSelect;
@end

@implementation MyDelegate
- (void)tableViewDidSelect {
    NSLog(@"Table view selected");
}
@end

// Set delegate
MyDelegate *del = [[MyDelegate alloc] init];
NSLog(@"Delegate set");
```
